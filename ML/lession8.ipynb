## Prepare the data

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.DataFrame(
    {
        "Batman": [5, 4, None, 1],
        "Spider Man": [4, 5, None, 2],
        "Finding Nemo": [1, None, 5, 4],
        "Toy Store": [None, 4, 2, 5],
    },
    index=["Ahmed", "Ali", "Sarah", "Mona"]
)
df
```

## Data cleaning

```python
df.isna().sum()
```

```python
df.fillna(0, inplace=True)
```

## Proceess the data
<div dir='rtl'>
    <p>لاحظنا ان العواميد تتكون من اسماء المشاهدين، ولكننا لا نريد ان نقارن المستخدمين، نريد ان نقارن الافلام لذالك نريد ان نعكس البيانات من <code>index</code> إلى <code>columns</code></p>
    <p>سوف نستخدم <code>T</code> وتعنى <code>Transpose</code> لقلب الجدول</p>
    <p></p>
</div>

```python
df.T
```

<p dir='rtl'>الان نريد ان نعرف نسبة التشابه لكل تقييم</p>

```python
similarity_metrix = cosine_similarity(df.T)
similarity_metrix
```

```python
dt = pd.DataFrame(similarity_metrix, index=df.columns, columns=df.columns)
dt
```

<p dir='rtl'>الان لنقم بعمل <code>function</code> صغيره لطباعة نسبة تشابه كل فيلم بناءً على اسم الفيلم</p>

```python
def system_recommend_matching(movie_name: str) -> None:
    print(f"since you liked {movie_name} then we recommend:")
    print("-" * 50)
    
    # show all rows but skip the selected movie
    recommendations = dt[movie_name].sort_values(ascending=False)[1:]
    
    for movie, score in recommendations.items():
        print(f"{movie} (Match: {score:.2%})")
```

```python
system_recommend_matching("Batman")
```

```python
movie = input("Enter the movie name that you have watched: ") # Spider Man

recommended_movie = dt[movie].sort_values(ascending=False)[1:] * 100
recommended_movie
```
