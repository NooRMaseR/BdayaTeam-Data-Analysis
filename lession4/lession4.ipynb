## Setup the Data

we will use `CSV` for this training, what is `CSV`?

`CSV` is a shortcut for `Comma Seprated Values`, to read a `CSV` file we will use `pd.read_csv`

<p dir='rtl'>
سوف نستخدم ملف <code>CSV</code> قى هذا التدريب, ما هو <code>CSV</code>؟
</p>
<p dir='rtl'>ال <code>CSV</code> هو اختصار لكلمة <code>Comma Seprated Values</code>, لقراءة ملف <code>CSV</code> سوف نستخدم <code>pd.read_csv</code></p>

```python
df = pd.read_csv(
    "e_commerce.csv",
    dtype={
        "order_id": np.int16,
        "quantity": np.int8,
        "shipping_days": np.int8,
    },
    true_values=['Yes'],
    false_values=['No'],
    index_col=0,
)
df['order_date'] = pd.to_datetime(df['order_date'])
df.head()
```

**Explaination**

* **dtype**: columns data types
* **true_values**: a `list` of values to convert into `True`
* **false_values**: a `list` of values to convert into `False`
* **index_col**: take the column index `0` to be the index of this `DataFrame`
* `df['order_date'] = pd.to_datetime(df['order_date'])`: used to convert a column into a `datetime` date type

<p dir='rtl'>
<strong>شرح</strong>
</p>

<ul dir='rtl'>
    <li>
        <strong>dtype</strong>: أنواع بيانات الأعمدة

    </li>

## Adding a column for item price after discount
we will use `insert` for this
<p dir='rtl'>
سوف نستخدم <code>insert</code> هنا
</p>

`insert(index, column_name, value)`

```python
df.head()
```

```python
unit_price_after_discount = np.round(df['unit_price'] * (1 - df['discount_pct']), 2)

df.insert(7, "unit_price_after_discount", unit_price_after_discount)
df.insert(8, "total_price_after_discount", unit_price_after_discount * df['quantity'])
```

```python
df.head()
```

### getting the total sales

```python
df['total_price_after_discount'].sum().round(2)
```
