## Data Types

### Integer Types: 
1. int8: 
    * (-128 to 127) 
    * size: 1 byte. 
2. int16: 
    * (-32,768 to 32,767)
    * size: 2 bytes. 
3. int32:
    * (-2,147,483,648 to 2,147,483,647)
    * size: 4 bytes. 
4. int64:
    * (-9,223,372,036,854,775,808 to 9,223,372,036,854,775,807)
    * size: 8 bytes.

### Unsigned Integer Types:
1. uint8: 
    * (0 to 255)
    * size: 1 byte.
2. uint16:
    * (0 to 65,535)
    * size: 2 bytes.
3. uint32:
    * (0 to 4,294,967,295)
    * size: 4 bytes.
4. uint64:
    * (0 to 18,446,744,073,709,551,615)
    * size: 8 bytes.

### Floating-Point Types: 
1. float16: 
    * Half-precision float
    * size: 2 bytes. 
2. float32: 
    * Single-precision float
    * size: 4 bytes. 
3. float64: 
    * Double-precision float
    * size: 8 bytes.


```python
rn = np.arange(0, 7000, 2, dtype=np.uint16)
rn
```

```python
candys = np.random.randint(10, 40, [3,3]).astype(np.int8)
candys
```

```python
print(f"maximum number: {candys.max()}")
print(f"minimum number: {candys.min()}")
print(f"total sum: {candys.sum()}")
print(f"total mean: {candys.mean()}")
```

## Axis
* axis = 0 -> columns -> عواميد
* axis = 1 -> rows -> صفوف

```python
print(f"maximum number per columns: {candys.max(axis=0)}")
print(f"maximum number per rows: {candys.max(axis=1)}")
```

```python
print(f"minimum number per columns: {candys.min(axis=0)}")
print(f"minimum number per rows: {candys.min(axis=1)}")
```

```python
print(f"total sum per columns: {candys.sum(axis=0)}")
print(f"total sum per rows: {candys.sum(axis=1)}")
```

```python
print(f"mean per columns: {candys.mean(axis=0)}")
print(f"mean per rows: {candys.mean(axis=1)}")
```
